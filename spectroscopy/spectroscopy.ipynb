```python
import chardet
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd

runs = [1, 2] 

fendas = [0.1, 0.2, 0.3, 0.5, 1.0]

for run in runs:
    for fenda in fendas:
        with open(f'data/hydrogen/h{fenda}mm{run}.txt', 'rb') as f:
            enc = chardet.detect(f.read())  # or readline if the file is large

        file = pd.read_csv(f'data/hydrogen/h{fenda}mm{run}.txt', header = 0, sep='\s+', encoding=enc['encoding'])

        fig1 = plt.figure(figsize=(10,6))

        axes = fig1.add_axes([0.1,0.1,0.8,0.8])

        axes.plot(file['posicao'], file["intensidade"], ls='-', label="Luz")

        axes.set_xlabel('Posio angular (rad)')
        axes.set_ylabel("Intensidade de luz")

        axes.set_title(f'Intensidade luminosa do hidrognio para a fenda de {fenda}mm')

        plt.grid(linestyle='-', linewidth=0.5)
        axes.legend(loc='upper right')

        plt.savefig(f'plotsHidrogenio/h{fenda}mm{run}.png', dpi=400)

```
